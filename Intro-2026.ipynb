```python
# Import all the things

import numpy as np
import matplotlib.pyplot as plt
import time

from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.visualization import plot_histogram

# This is "magic" that lets Qiskit draw into notebook cells
%matplotlib inline  
```

## Support for running circuits and showing results

```python
def run_project(make_gates, num_qubits, num_shots=1):
    # Build the quantum circuit. 
    
    # First, make the data structure to hold the circuit.
    # We assume there will be 1 classical bit for each quantum bit -
    # the arguments to QuantumCircuit are # qubits, # classical bits.
    qc = QuantumCircuit(num_qubits, num_qubits) 

    # Now insert the gates for this circuit
    make_gates(qc)

    # Finally, measure each qubit into its corresponding classical bit 
    for line_num in range(num_qubits):
        qc.measure(line_num, line_num)

    # Get the simulator and transpile the circuit for simulation
    simulator = Aer.get_backend('qasm_simulator')
    compiled_circuit = transpile(qc, simulator)

    # run the simulator and give the outputs convenient names
    job = simulator.run(compiled_circuit, shots=num_shots, seed_simulator=int(time.time()))
    result = job.result()
    counts = result.get_counts() 

    # draw the circuit, state the measurements, and show a bar graph
    circuit_fig = qc.draw(output='mpl', cregbundle=False, style='iqp', initial_state=True)
    circuit_fig.savefig('circuit.png', dpi=300, bbox_inches='tight')
    plt.show()
    plt.close(circuit_fig)  

    print('measurements:')
    for bitstring, count in sorted(counts.items(), key=lambda x: -x[1]):
        print(f"{bitstring}: {count} times (percentage {count/num_shots:.4f})")
    plot_histogram(counts, figsize=[3, 2])

    ax = plt.gca()    
    max_count = 0
    for key, value in counts.items():
        max_count = max(max_count, value) 

    ax.set_yticks([0, max_count])
    plt.xlabel('Measured States', fontsize=14)
    plt.ylabel('Counts', fontsize=14)

    plt.savefig('histo.png', dpi=300, bbox_inches='tight')
    plt.show()
```
